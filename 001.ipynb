```python
import mysql.connector as mc
from mysql.connector import Error
```

```python
def insert_data(conn):
    # 데이터 삽입
    sql = 'INSERT INTO customers ' \
    '(name, email, phone, point, grade, join_date, last_purchase_date) ' \
    'VALUES(%s,%s,%s,%s,%s,%s,%s)'

    values = ('홍길동', 'hong2.cs@email.com', '010-1111-0001', 5000, 'Gold', '2023-01-15', '2024-02-20')

    cursor = conn.cursor()
    cursor.execute(sql, values)    
    conn.commit()
```

```python
conn = None
try:    
    conn =  mc.connect(
        host = 'localhost',  # DB 주소(127.0.0.1)
        user = 'root',
        password = 'root1234!',
        database = 'shop_db'
    )
    if conn.is_connected:
        print('접속성공')     
    insert_data(conn)
except Error as e:
    print(f'데이터베이스 오류 : {e}')   
finally:
    if conn:
        conn.close()        
        print('연결 해제')
```
