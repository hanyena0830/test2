```python
import pandas as pd
d = {'category': ['전자제품', '식품','의류'], 'value': [100, 80, 50]}
df = pd.DataFrame(d)
df
```

```python
# 간단하고 빠른 차트 - 일반적인 차트에 적합
import plotly.express as px
fig = px.bar(df, x='category', y='value')
fig.show()
```

```python
import plotly.graph_objects as go
# 세밀한 제어
fig = go.Figure()
fig.add_trace(go.Bar(x=['a','b'], y=[10,20]))
fig.add_trace(go.Bar(x=['a','b'], y=[100,200]))
fig.update_layout(title='Custom Chart')
fig.show()
```

# 파이차트

비율, 구성요소들 보여줄때

```python
import plotly.express as px
import pandas as pd
df = pd.DataFrame({
    'grade' : ['gold','silver','bronze'],
    'count' : [30,50,10]
})
# 파이차트생성
fig = px.pie(df, values='count',names='grade',title='고객 등급 분포',
             color_discrete_sequence=px.colors.sequential.amp)
fig.show()
```

# 막대그래프

비교, 순위 표시

```python
import plotly.express as px
import pandas as pd
# 샘플데이터
df = pd.DataFrame({
    'category' : ['전자기기','의류','가방','생활용품'],
    'count' : [4,3,2,1]
})

# 막대 그래프
fig = px.bar(df, x='category', y='count',
       title = '카테고리별 상품 수',
       color = 'count',  # 색상(값에따라),
       color_continuous_scale='Blues'  # 색상 그라데이션    
)
fig.show()
```

# 그룹화된 막대 그래프

```python
import plotly.express as px
import pandas as pd
# 샘플데이터
df = pd.DataFrame({
    'category' : ['A','A','B','B'],
    'type' : ['Type1','Type2','Type1','Type2'],
    'value' : [10,18,8,12]
})

# 막대 그래프
fig = px.bar(df, x='category', y='value',       
       color = 'type', 
       barmode='group',
       color_continuous_scale='Blues'  # 색상 그라데이션    
)
fig.show()
```

```python
# 수평 막대 그래프
fig = px.bar(df, x='value', y='category',orientation='h')
fig.show()
```

# 선그래프

시간흐름, 추세 분석  x축이 시계열

```python
import plotly.express as px
import pandas as pd

# 샘플 데이터
df = pd.DataFrame({
    'month': ['2024-01', '2024-02', '2024-03', '2024-04'],
    'new_customers': [15, 23, 18, 30]
})

# 선 그래프
fig = px.line(
    df,
    x='month',
    y='new_customers',
    title='월별 신규 고객 수',
    markers=True  # 데이터 포인트에 마커 표시
)

fig.show()
```

# 영역그래프

누적 값, 전체양의 변화 강조

```python
import plotly.express as px
import pandas as pd

# 샘플 데이터
df = pd.DataFrame({
    'order_day': ['2024-02-15', '2024-02-16', '2024-02-17', '2024-02-18'],
    'daily_revenue': [1235000, 375000, 143000, 1550000]
})

# 영역 그래프
fig = px.area(
    df,
    x='order_day',
    y='daily_revenue',
    title='일별 매출 추이'
)

fig.show()
```
